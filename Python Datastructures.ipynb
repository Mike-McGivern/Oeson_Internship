#Lists

- Created within the [] bracket
- output is within []
- Element has a fixed index value
- Access / Updating / Modifying / Adding / Deleting
- Slicing
- Mutable
- can have a mixture of data types "Heterognous elements"


Examples: Shopping List, Task Management
    

```python
#Create a list
myList = ['meeting', 'shopping', 'movie', 27, 45, 63, 89]
print(myList)

#Access an element
print(myList[2])

#Slicing

print(myList[2:6])
print(myList[-5::])
print(myList[-1::])
print(myList[::-1])
print(myList[0:7:2])

#add element
myList.append('learning')
print(myList)

#add at specific position (Insert)
myList.insert(3, 44)
print(myList)

#remove element

myList.remove('learning')
print(myList)
myList.remove(44)
print(myList)

#pop function
myList.pop(2)
print(myList)


print(len(myList))

```

```python
#Create a Dictionary

myDict = {
    'a' : 1,
    'b' : 2,
    'c' : 3,
    'd' : 3,
    'e' : 2,
    'color' : 'red'
}
print(myDict)

#Accesss elements

print(myDict['b'])

print(myDict['color'])

print(myDict.get('f'))
print(myDict.get('f', 'N/A'))

print(list(myDict.keys()))
print(list(myDict.items()))
print('b' in myDict)

#Add or Update
myDict['f'] = 'blue'
print(myDict)

myDict['f'] = 4
print(myDict)

#Remove
deleted = myDict.pop('f')
print(deleted)
print(myDict)
del myDict['color']
print(myDict)

myDict.clear()
print(myDict)
```
